```python
import pandas as pd

input_csv = "Saved_Data/all_brands_clean.csv"
output_csv = "Saved_Data/fragrances_with_nan.csv"

df = pd.read_csv(input_csv)

nan_per_column = df.isna().sum()
total_nan_values = int(nan_per_column.sum())

print(f"Total NaN values in dataset: {total_nan_values}")
print("\nNaN values per column:")
print(nan_per_column[nan_per_column > 0])

fragrances_with_nan = df[df.isna().any(axis=1)].copy()
print(f"\nNumber of fragrance rows with at least one NaN: {len(fragrances_with_nan)}")

fragrances_with_nan.to_csv(output_csv, index=False)
print(f"Saved fragrances with NaN values to: {output_csv}")
```
